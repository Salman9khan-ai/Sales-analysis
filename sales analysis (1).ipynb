```python
import pandas as pd
import statsmodels.api as sm

# Sample data
data = {
    'Category': ['Phones', 'Printers', 'Electronic Games', 'Bookcases', 'Chairs'],
    'Sales Quantities': [4, 8, 7, 8, 14],
    'Profits': [1151, 712, 658, 146, 64],
    'Profit Margins (%)': [28, 21, 15, 5, 2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Prepare the independent variables (X) and dependent variable (y)
X = df[['Sales Quantities', 'Profit Margins (%)']]  # Predictor variables
X = sm.add_constant(X)  # Add constant for intercept
y = df['Profits']  # Target variable

# Fit the regression model
model = sm.OLS(y, X).fit()

# Display the summary
print(model.summary())
```

```python
Here’s an interpretation of your regression analysis results:

---

### **1. Model Fit (R-squared and Adj. R-squared)**
- **R-squared (0.976):** This indicates that 97.6% of the variance in profits is explained by the predictors (Sales Quantities and Profit Margins). The model is highly effective in predicting profits based on these variables.
- **Adjusted R-squared (0.951):** Adjusted for the number of predictors, this still shows a very strong fit. It means the model generalizes well to the data.

---

### **2. F-statistic and Prob (F-statistic)**
- **F-statistic (39.84) and Prob (F-statistic: 0.0245):** The low p-value (< 0.05) indicates that the model as a whole is statistically significant. This suggests that Sales Quantities and Profit Margins are effective predictors of profits.

---

### **3. Coefficients and Their Interpretation**
- **Intercept (91.35):** When both sales quantities and profit margins are zero, the expected profit would be approximately 91.35. While the intercept is not directly actionable, it sets the baseline for the equation.

- **Sales Quantities Coefficient (-10.09):**
  - A **negative coefficient** suggests that increasing sales quantities by 1 unit is associated with a decrease in profits by 10.09 units, assuming profit margins remain constant.
  - This aligns with your correlation analysis, where bulk sales are tied to low-margin items. Increasing low-margin product sales negatively impacts overall profits.

- **Profit Margins Coefficient (37.86):**
  - A **positive coefficient** implies that a 1% increase in profit margins is associated with an increase in profits by 37.86 units, holding sales quantities constant.
  - This shows the critical importance of profit margins in driving profitability.

---

### **4. Statistical Significance (P-values)**
- **Sales Quantities (P-value: 0.712):**
  - A high p-value (> 0.05) indicates that sales quantities are **not statistically significant** in predicting profits. This suggests that focusing solely on increasing sales volume may not yield substantial profit growth.

- **Profit Margins (P-value: 0.041):**
  - A low p-value (< 0.05) shows that profit margins are **statistically significant**. They strongly influence profitability and should be a priority in decision-making.

---

### **5. Optimization Suggestions**
Based on the results:
1. **Prioritize Profit Margins:**
   - Focus on improving margins for all product categories, especially high-volume ones like Chairs and Bookcases.
   - Strategies could include reducing costs, introducing premium versions, or slightly increasing prices.

2. **Limit Reliance on Low-Margin, High-Sales Items:**
   - Bulk sales (e.g., Chairs) may dilute profitability. Consider:
     - Cost efficiency measures.
     - Bundling low-margin items with high-margin products to improve overall profit.

3. **Emphasize High-Margin Items:**
   - Categories like Phones and Printers have strong margins and profitability. Allocate marketing and sales efforts toward these categories.

4. **Introduce Products with Higher Margins:**
   - Explore expanding the product portfolio to include items similar to Phones, which have high profit margins even with lower sales quantities.

5. **Data-Driven Adjustments:**
   - Use regression insights to fine-tune pricing and promotion strategies dynamically, based on real-time sales and profit margin performance.

---

```

```python
#SCENARIO SIMULATION
import pandas as pd

# Original data
data = {
    'Category': ['Phones', 'Printers', 'Electronic Games', 'Bookcases', 'Chairs'],
    'Sales Quantities': [4, 8, 7, 8, 14],
    'Profits': [1151, 712, 658, 146, 64],
    'Profit Margins (%)': [28, 21, 15, 5, 2]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Define a function for scenario simulation
def simulate_profit(data, sales_multiplier, margin_increase):
    # Simulate new sales quantities
    new_sales = [x * sales_multiplier for x in data['Sales Quantities']]
    
    # Simulate new profit margins
    new_margins = [x + margin_increase for x in data['Profit Margins (%)']]
    
    # Calculate new profits based on simulated values
    new_profits = [
        new_sales[i] * new_margins[i] * 10  # Assume profits scale with sales and margin (scaled by 10 for simplicity)
        for i in range(len(data['Category']))
    ]
    
    # Create a simulated DataFrame
    simulated_data = pd.DataFrame({
        'Category': data['Category'],
        'Simulated Sales Quantities': new_sales,
        'Simulated Profit Margins (%)': new_margins,
        'Simulated Profits': new_profits
    })
    
    return simulated_data

# Run scenario simulation (e.g., increasing sales by 1.5x and profit margins by 5%)
sales_multiplier = 1.5
margin_increase = 5  # Percentage increase
simulated_results = simulate_profit(data, sales_multiplier, margin_increase)

# Display simulated results
print("Scenario Simulation Results:")
print(simulated_results)
```

```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px

# Sample data
data = {
    'Category': ['Phones', 'Printers', 'Electronic Games', 'Bookcases', 'Chairs'],
    'Sales Quantities': [4, 8, 7, 8, 14],
    'Profits': [1151, 712, 658, 146, 64],
    'Profit Margins (%)': [28, 21, 15, 5, 2]
}
df = pd.DataFrame(data)

# Initialize Dash app
app = Dash(__name__)

# Create a plotly scatter plot
fig = px.scatter(
    df, 
    x='Sales Quantities', 
    y='Profits', 
    size='Profit Margins (%)', 
    color='Category', 
    title='Profit vs Sales Quantities by Category',
    labels={'Sales Quantities': 'Sales Volume', 'Profits': 'Profit'}
)

# Define app layout
app.layout = html.Div([
    html.H1("Interactive Dashboard", style={'textAlign': 'center'}),
    dcc.Graph(figure=fig),  # Render the plotly graph
    html.P("Explore sales and profitability data interactively!")
])

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```

```python
from dash import Dash, html, dcc
import pandas as pd
import plotly.express as px

# Sample data
data = {
    'Category': ['Phones', 'Printers', 'Electronic Games', 'Bookcases', 'Chairs'],
    'Sales Quantities': [4, 8, 7, 8, 14],
    'Profits': [1151, 712, 658, 146, 64],
    'Profit Margins (%)': [28, 21, 15, 5, 2]
}
df = pd.DataFrame(data)

# Initialize Dash app
app = Dash(__name__)

# Create a plotly scatter plot with enhanced visuals
fig = px.scatter(
    df, 
    x='Sales Quantities', 
    y='Profits', 
    size='Profit Margins (%)', 
    color='Category', 
    title='Profit vs Sales Quantities by Category',
    labels={'Sales Quantities': 'Sales Volume', 'Profits': 'Profit'},
    opacity=0.8  # Adjust dot opacity for better contrast
)
fig.update_layout(
    paper_bgcolor='#f0f8ff',  # Set a light background color (named CSS color or hex)
    plot_bgcolor='#ffffff',  # Set a solid white background for the chart area
    title_font=dict(size=20, color='darkblue', style='italic'),  # Italic title font
    font=dict(family='Arial, Italic', size=12)  # Global italic font
)
fig.update_traces(
    marker=dict(line=dict(width=2, color='black'))  # Add bold borders around dots for contrast
)

# Define app layout with simulated fading background using HTML/CSS
app.layout = html.Div([
    html.H1("Interactive Dashboard", style={'textAlign': 'center', 'fontStyle': 'italic', 'color': '#4B0082'}),
    dcc.Graph(figure=fig),  # Render the plotly graph
    html.P("Explore sales and profitability data interactively!", style={'fontStyle': 'italic', 'textAlign': 'center', 'color': '#8A2BE2'})
], style={'background': 'linear-gradient(to bottom, #ff9a9e, #fad0c4, #fbc2eb)', 'padding': '20px'})  # Fading background added via CSS

# Run the app
if __name__ == '__main__':
    app.run(debug=True)
```
